First we have to find which columns we have to replace or drop .


```python
#1. First drop rows of Creator name  and no. of playlist because we can not replace it with other value.
```

```python
#DROPING NaN value row 
#df1.drop_duplicates(['B'], inplace=True) or df.drop_duplicates('B', inplace=True)
df1=df1.dropna(subset=['creator_name','no_of_playlist','subtitle','no_of_likes'])
df1.isna().sum()
```

```python
#2 Checking data of creator_gender
df1.creator_gender.value_counts(dropna =False)
```

From above, we find that thier are one other parameter which is company and we can define a gender of any comapany and industry so lets check the creator name of NaN value .

```python
df1[df1.creator_gender.isnull()].creator_name.unique()
```

```python
#We can not replace null value with mode 
#Changing NaN value to Others
df1.loc[df1.creator_gender.isnull(),'creator_gender'] = 'others'

df1.creator_gender.value_counts(dropna =False)
```

```python
#Changing company parameter to others
df1['creator_gender'] = df1['creator_gender'].map({'Company':'others','Male':'Male','Female':"Female","others":"others"})

df1.creator_gender.value_counts(dropna =False)
```

```python
#3.Now check the boxplot of number of likes  
df1.no_of_likes.plot(kind='box')
```

```python
# Replacing NaN value with median
df1["no_of_likes"].fillna(int(df1['no_of_likes'].median()),inplace=True) 
```

```python
#4.  language_of_the_video
df1.language_of_the_video.value_counts(dropna=False).plot(kind='bar')
```

```python
#Too many different language are there . We can not replace it with mode it will skiw the data better to drop these rows
df1.dropna(subset=['language_of_the_video'],inplace=True)
df1.isna().sum()
```

```python
#Replacing NaN with mode of the Categorical Variable
df1.date_of_the_last_comment.fillna(df1.date_of_the_last_comment.mode()[0],inplace = True)
df1.isna().sum()
```
